```python
# %matplotlib inline을 설정하면 matplotlib.pyplot의 show 함수를 호출하지 않아도
# 주피터 노트북 안에서 그래프가 표시됨
%matplotlib inline
import pandas as pd                # 데이터를 저장하고 처리하는 패키지
import matplotlib as mpl           # 그래프를 그리는 패키지
import matplotlib.pyplot as plt    # 그래프를 그리는 패키지

# csv 파일을 읽어서 DataFrame 객체로 만듦. 인덱스 컬럼은 point로 설정
#df = pd.read_csv('진보와보수.csv', index_col='point') 
#오류 발생: UnicodeDecodeError: 'utf-8' codec can't decode byte 0xbc in position 0: invalid start byte
# csv 파일을 읽어서 DataFrame 객체로 만듦. 인덱스 컬럼은 point로 설정
df = pd.read_csv('에타_최소최대마일리지_정원 데이터.csv', header = 0)
```

```python
df
```

```python
#new column: 에브리타임 인원/수원 정원
df['et/capacity'] = df['et_applicants']/df['course capacity']
df
```

```python
plt.scatter(df['min_mileage'], df['et/capacity'])
```

```python
plt.scatter(df['max_mileage'], df['et/capacity'])
```

```python
df['min_mileage'].plot(color='#ff0000')
plt.grid()
plt.legend()
plt.title("minimum mileage")
plt.xlabel("et/capacity")
plt.ylabel("min_mileage")
plt.show()
```

```python
df['max_mileage'].plot(color='#0067a3')
plt.grid()
plt.legend()
plt.title("maximum mileage")
plt.xlabel("et/capacity")
plt.ylabel("max_mileage")
plt.show()
```
